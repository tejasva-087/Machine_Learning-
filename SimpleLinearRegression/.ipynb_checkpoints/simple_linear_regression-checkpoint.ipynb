```python
import matplotlib.pyplot as plt
import numpy as np
import pylab as pl  # This is a sub module of matplotlib
from sklearn import linear_model  # Importing the regression model

# Due to some issue with pandas so that's why we are using this warnings and all
import warnings
warnings.filterwarnings("ignore", "\nPyarrow", DeprecationWarning)
import pandas as pd


# This line below is used in jupyternotebook to view graphs
%matplotlib inline
```

```python
df.head()  # Looking at the dataset gives the column reading upto 5 columns
```

```python
# This gives the summary of the code
df.describe()  # Description of data
```

```python
cdf = df[['ENGINESIZE','CYLINDERS','FUELCONSUMPTION_COMB','CO2EMISSIONS']]
cdf.head(9)
```

```python
# Plotting the graph for the selected data
cdf.hist()
plt.show()
```

```python
plt.scatter(cdf.FUELCONSUMPTION_COMB, cdf.CO2EMISSIONS,  color='green')
plt.xlabel("FULECONSUMPTION_COMBO")
plt.ylabel("CO2EMMISIONS")
plt.show()
```

```python
plt.scatter(cdf.ENGINESIZE, cdf.CO2EMISSIONS,  color='red')
plt.xlabel("Engine size")
plt.ylabel("Emission")
plt.show()
```

```python
plt.scatter(cdf.CYLINDERS, cdf.CO2EMISSIONS, color="purple")
plt.xlabel("CYLINDER")
plt.ylabel("CO2EMMISIONS")
plt.show()
```

```python
# Applying the train/test split
msk = np.random.rand(len(df)) < 0.8
# Training data
train = cdf[msk]
# Testing data
test = cdf[~msk]  # ~ Seperated the False data generated by the msk
```
